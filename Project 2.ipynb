# Statistical Analysis

**Anova Test**

```python
from scipy import stats

# Contains the columns 'Age' and 'Gender'
male_ages = df_copy[df_copy['Vict Sex'] == 'M']['Vict Age']
female_ages = df_copy[df_copy['Vict Sex'] == 'F']['Vict Age']
transgender_ages = df_copy[df_copy['Vict Sex'] == 'X']['Vict Age']

# Calculating the p-value and t-statistic 
t_stat, p_value = stats.ttest_ind(male_ages, female_ages)

print(f"T-statistic: {t_stat}, P-value: {p_value}\n")

# Conclusion based on p-value
if p_value < 0.05:
    print("The results of the t-test indicate that there is a statistically significant difference in the average age between males and females (p-value < 0.05).")
    print("Therefore, we reject the null hypothesis and conclude that the mean age of males is different from the mean age of females in the dataset.")
else:
    print("The results of the t-test show that there is no statistically significant difference in the average age between males and females (p-value >= 0.05).")
    print("Thus, we fail to reject the null hypothesis, and we conclude that the mean age of males and females in the dataset is similar.")
```

**z-Test**

```python
import numpy as np
import scipy.stats as stats

# Assuming df_copy contains the column 'Vict Age' for age analysis
sample_ages = df_copy['Vict Age']

# Calculating sample mean and sample standard deviation
mean_age = np.mean(sample_ages)
std_dev_age = np.std(sample_ages, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimate)
n = len(sample_ages)  # Sample size

# Assume a population mean (e.g., for example purposes, we assume 30 years)
population_mean = df_copy['Vict Age'].mean()

# Step 3: Calculate the Z-statistic
z_stat = (mean_age - population_mean) / (std_dev_age / np.sqrt(n))

# Step 4: Compute the p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))  # Two-tailed p-value

# Output Z-statistic and p-value
print(f"Z-statistic: {z_stat}, P-value: {p_value}")

# Conclusion based on p-value
if p_value < 0.05:
    print("\nWe reject the null hypothesis. There is a statistically significant difference between the sample mean age and the population mean.")
else:
    print("\nWe fail to reject the null hypothesis. There is no statistically significant difference between the sample mean age and the population mean.")
```
